```python
tips = pd.read_csv("https://raw.githubusercontent.com/schloerke/cranvasOLD/master/files/data/tips.csv")
```

```python
tips
```

```python
tips['tip_percapita'] = tips['tip']/tips['size']
tips['tip_percent'] = tips['tip']/tips['total_bill']
```

```python
#tips (dataframe que armazena os dados) com novas colunas
tips
```

```python
tip = tips['tip']
sns.distplot(tip)

#skewness - assmimetria/obliquidade
skewness = scipy.stats.skew(tip)
print(skewness)
```

```python
tip_percapta = tips['tip_percapita']
sns.distplot(tip_percapta)

skewness = scipy.stats.skew(tip_percapta)
print(skewness)
```

```python
tip_percent = tips['tip_percent']
sns.distplot(tip_percent)

skewness = scipy.stats.skew(tip_percent)
print(skewness)
```

```python
bmi = pima['bmi']
bmi.mean() 
```

```python
#Remove NaNs
pima_sem_nan = pima.dropna()
pima_sem_nan
```

```python
bmi_sem_nan = pima_sem_nan['bmi']
bmi_sem_nan.mean()
```

```python
#Substitui NaNs pelas médias
pima_media = pima.fillna(pima.median())
pima_media
```

```python
diastolic_bp = pima_media['diastolic_bp']
diastolic_bp.std() #valor do desvio padrão da coluna 'diastolic_bp' depois da substituição dos NaNs pelas médias'
```
